```python
import numpy as np
import cv2

from matplotlib import pyplot as plt
```

```python
img = cv2.imread('samples/radar_rnd.png')
```

```python
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
COLORS = [
    [51, 51, 51],
    [155, 85, 197],
    [232, 10, 231],
    [170, 11, 16],
    [202, 1, 7],
    [255, 1, 1],
    [255, 142, 4],
    [230, 191, 2],
    [245, 253, 9],
    [9, 141, 4],
    [6, 196, 0],
    [16, 242, 10],
    [22, 0, 205],
    [12, 147, 255],
    [21, 224, 238]
]
```

```python
rgbimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
g1 = np.array([16, 242, 10])
g2 = np.array([16, 242, 10])

mask = cv2.inRange(rgbimg, np.array(COLORS[0]), np.array(COLORS[0]))
for c in COLORS[1:]:
    mask = mask | cv2.inRange(rgbimg, np.array(c), np.array(c))

res = cv2.bitwise_and(rgbimg, rgbimg, mask=mask)
```

```python
plt.imshow(res)
plt.show()
```

```python
bgrimg = cv2.cvtColor(res, cv2.COLOR_RGB2BGR)
cv2.imwrite('samples/radar_rnd_clean.png', bgrimg)
```
